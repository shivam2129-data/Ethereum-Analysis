```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('transactions-final.csv')
df_avg=pd.read_csv('transactions-avg.csv')
```

```python
df1=df.T
df1_avg=df_avg.T
```

```python
k=0
dfn=pd.DataFrame([])
for i in df1.index:
  if k%2==0:
    dfn.loc[k,0]=i
  else:
    dfn.loc[k-1,1]=i
  k+=1
k=0
dfn_avg=pd.DataFrame([])
for i in df1_avg.index:
  if k%2==0:
    dfn_avg.loc[k,0]=i
  else:
    dfn_avg.loc[k-1,1]=i
  k+=1
```

```python
dfn.head()
```

```python
dfn = dfn.applymap(lambda x: str(x).replace('[["', '').replace('"]]', '').replace('["','').replace('"','').replace(']',''))
dfn_avg = dfn_avg.applymap(lambda x: str(x).replace('[["', '').replace('"]]', '').replace('["','').replace('"','').replace(']',''))
```

```python
dfn.columns=['Year-Month','Transaction']
dfn_avg.columns=['Year-Month','Transaction']
dfn['Transaction']=dfn['Transaction'].astype('int')
dfn['Year-Month']=pd.to_datetime(dfn['Year-Month']).dt.to_period('M')
dfn=dfn.sort_values(by='Year-Month')
dfn_avg['Transaction']=dfn_avg['Transaction'].astype('float')
dfn_avg['Year-Month']=pd.to_datetime(dfn_avg['Year-Month']).dt.to_period('M')
dfn_avg=dfn_avg.sort_values(by='Year-Month')
```
